## Machine learning

```python
X, y = test1_master_ready.iloc[:, 3:], test1_master_ready.iloc[:, 2]
y
```

```python
# LeaveOneGroupOut
cv = LeaveOneGroupOut()
groups = test1_master_ready['patient_ID'].values
kbest = SelectKBest(score_func=f_classif, k=100)
estimator0 = XGBClassifier(random_state=0)
pipe = Pipeline([('prep-1',kbest), ('estimator', estimator0)])
result = cross_validate(pipe, X.values, y.values, groups=groups, cv=cv, scoring='balanced_accuracy', verbose=10)
result, np.mean(result['test_score'])
```

```python
# random split
cv = StratifiedKFold(5, shuffle=True, random_state=0)
#groups = test3_master_ready.iloc[:, 1].values
kbest = SelectKBest(score_func=f_classif, k=100)
estimator0 = XGBClassifier(random_state=0)
pipe = Pipeline([('prep-1',kbest), ('estimator', estimator0)])
result = cross_validate(pipe, X.values, y.values, groups=None, cv=cv, scoring='balanced_accuracy', verbose=10)
result, np.mean(result['test_score'])
```

```python
# LeaveOneGroupOut
cv = LeaveOneGroupOut()
groups = test1_master_ready['patient_ID'].values
kbest = SelectKBest(score_func=f_classif, k=100)
estimator0 = XGBClassifier(random_state=0)
pipe = Pipeline([('prep-1',kbest), ('estimator', estimator0)])
predicted = cross_val_predict(pipe, X.values, y.values, groups=groups, cv=cv, verbose=10)
predicted
```
